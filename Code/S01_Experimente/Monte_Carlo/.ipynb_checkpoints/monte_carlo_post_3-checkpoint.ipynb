# Einlesen der Daten

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#from Code.S02_Allgemein.process_hdf5 import plot_from_exp_id_onedist

plt.rcParams['figure.figsize'] = [8, 6] # Größe der Figure festsetzen
```

```python
path = r"C:\Users\timbr\OneDrive\Masterarbeit\Daten\2020_09_13_Monte_Carlo_3\\"
h5_filename1 = "2020_09_14__14_14_46_Monte_Carlo_Data.hdf5"
filename1 = "2020_09_14__14_14_46_Monte_Carlo_DataFrame.pkl"
df_raw = pd.read_pickle(path+filename1)

df_raw
```

```python
df_raw.columns
```

WICHTIG:
Die Metrik ``disq`` für dieses Experiment ignorieren! Die Metrik schlägt bei 200s an, obwohl es in diesem Experiment keine zweite Störung bei 200s gab.

```python
df_raw['dist_len1'] = [i[-1] for i in df_raw['dist_name1'].str.split('_')]
df_raw['dist_type'] = [i[0]+i[1] for i in df_raw['dist_name1'].str.split('_')] 

# in numerische Spalten umwandeln
numeric_columns = ['resloss_sum', 'overlap', 'dist_len1', 'rt_noise1']

for i in numeric_columns:
    df_raw[i] = df_raw[i].apply(pd.to_numeric, errors='coerce')

df_raw['res_triag'] = df_raw['resloss_sum']/20

df_raw.drop(['A', 'deadtime', 'dist_id', 'model_qual', 'q', 'signal_power', 'sigma', 'react_mode', 'dist_name', 'noise_sigma', 'detect_mode', 'resloss_sum', 'disq'], axis=1, inplace=True)
print(df_raw.shape)
```

```python
df = df_raw
```

```python
df_group = df[['dist_len1', 'overlap', 'res_triag', 'rt_noise1']].groupby(['dist_len1']).mean() # nach Störlange gruppieren
df_group
```

```python
df2 = df[df['dist_type'] == 'cosinus2']
df2[['dist_len1', 'overlap', 'res_triag']].groupby(['dist_len1']).mean()
```

```python
df_group.to_csv("MC3_overview.csv")
```

# Plot Überlappung bzw. Resilience Triangle über Dauer der Störung

```python
# in Listen abspeichern

dist_len = []
overlap = []
resloss = []
rt = []

for i in df_group.iterrows():
    dist_len.append(i[0])

for i in df_group['overlap']:
    overlap.append(i)

for i in df_group['res_triag']:
    resloss.append(i)
    
for i in df_group['rt_noise1']:
    rt.append(i)

# Plots erzeugen
    
fig, ax = plt.subplots(2, 1, sharex=True, figsize=(8, 4.5))   
plt.subplots_adjust(hspace = 0.2)

for i in range(0,2):
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].yaxis.set_ticks_position('left')
    ax[i].xaxis.set_ticks_position('bottom')
    ax[i].yaxis.set_tick_params(direction="out", width=1)
    ax[i].xaxis.set_tick_params(direction="out", width=1)
    
    for tick in ax[i].yaxis.get_major_ticks():
        tick.label.set_fontsize(11)
    for tick in ax[i].xaxis.get_major_ticks():
        tick.label.set_fontsize(11)

ax[0].plot(dist_len, overlap, 'x-')
ax[0].set(ylabel = "Überlappung")
ax[1].plot(dist_len, rt, 'x-')
#ax[1].set(ylabel = "Resilience Triangle")
#ax[2].plot(dist_len, rt, 'x-')
ax[1].set(ylabel = "rt")
ax[1].set(xlabel = "Stördauer in s")


plt.savefig('MC3.svg')
plt.show()
```
