## Modellqualität

Bei den modelbasierten Methoden wird parallel zur eigentlichen Simulation eine Simulation ohne den Eingang der Störung mitsimuliert. Weichen beide Modell signifikant ab, kann man auf eine Störung schließen. In der Realität verfügt man über kein exaktes Modell und daher soll hier gezeigt werden, wie sich die Modellqualität auf die modellbasierten Methoden auswirkt.

Wie man in der folgenden Tabelle sieht, ist der ``false_pos_count`` bei allen Methoden für das nichtexakte Modell deutlich höher als für dass exakte Modell. Fasst man alle Methoden zusammen, ist erkenntlich, dass der ``false_pos_count`` für das nicht exakte Modell ca. doppelt so groß ist. Auch der ``false_neg_count`` sinkt, wenn die Modellqualität abnimmt. Es lässt sich damit festhalten, dass die Modellqualität für die Erkennung essentiell ist.

```python
df_model = df.loc[ (df['detect_mode'] != 'gradient algo')] # & (df['noise_sigma'] == 0.01) ]
df_model[['detect_mode',
       'model_qual', 'false_pos_count', 'false_neg_count',
       'time_to_detect', 'time_to_recover', 'yexceed+5%', 'resloss_sum',
       'max_loss']].groupby(['detect_mode', 'model_qual']).mean()
```

# Vergleich der Methoden bei gleicher Störung

Im folgenden Teil wird verglichen, wie die Methoden auf die gleiche Störung reagieren.

```python
# list mit allen Störungen
com_sel = df.groupby(['dist_id']).max()

dist_list = []
for i in com_sel.dist_list:
    dist_list.append(i)
```

Erzeugen eines Dataframes mit dem gerinsten ``res_loss_sum`` für alle Störungen.

```python
flag = 0

for i in dist_list:
    df_dist = df.loc[df['dist_list'] == i]
    df_dist.sort_values(by=['resloss_sum'], inplace = True)
    df_new = df_dist.iloc[0:1]
    
    if flag == 0:
        df_best = df_new
        flag = 1
    else:
        df_best = pd.concat([df_best, df_new])
        
df_best
```

Im Folgenden wurde für jede Störung die beste Methode hinsichtlich einer Metrik ermittelt. In der folgenden Tabelle sieht man nun , wie oft eine Methode die beste Methode war.

```python
metrices = ['time_to_recover', 'yexceed+5%', 'resloss_sum',
       'max_loss']
# 'false_pos_count', 'false_neg_count', 'time_to_detect' werden ignoriert, da hier das Minimum oft nicht eindeutig ist

methods = df['detect_mode'].unique()

# new DataFrame
df_count = pd.DataFrame(columns=['methods'] + metrices)

for i in range(0, len(methods)):
    
    df_count.loc[i] = [methods[i]] + [0]*len(metrices)

df_count.set_index('methods', inplace=True)

for matrix in metrices:
    
    flag = 0
    for dist in dist_list:
        df_dist = df.loc[df['dist_list'] == dist]
        df_dist.sort_values(by=[matrix], inplace = True)
        df_new = df_dist.iloc[0:1]

        if flag == 0:
            df_best = df_new
            flag = 1
        else:
            df_best = pd.concat([df_best, df_new])
            
    result = df_best.detect_mode.value_counts()
            
    for i in range(0, len(result)):
        df_count.loc[result.index[i], matrix] = result[i]
    
df_count["sum"] = df_count.sum(axis=1)

df_count
```

```python
metrices = ['time_to_recover', 'yexceed+5%', 'resloss_sum',
       'max_loss']
# 'false_pos_count', 'false_neg_count', 'time_to_detect' werden ignoriert, da hier das Minimum oft nicht eindeutig ist

df = df[df['false_pos_count']==0]

methods = df['detect_mode'].unique()

# new DataFrame
df_count = pd.DataFrame(columns=['methods'] + metrices)

for i in range(0, len(methods)):
    
    df_count.loc[i] = [methods[i]] + [0]*len(metrices)

df_count.set_index('methods', inplace=True)

for matrix in metrices:
    
    flag = 0
    for dist in dist_list:
        df_dist = df.loc[df['dist_list'] == dist]
        df_dist.sort_values(by=[matrix], inplace = True)
        df_new = df_dist.iloc[0:1]

        if flag == 0:
            df_best = df_new
            flag = 1
        else:
            df_best = pd.concat([df_best, df_new])
            
    result = df_best.detect_mode.value_counts()
            
    for i in range(0, len(result)):
        df_count.loc[result.index[i], matrix] = result[i]
    
df_count["sum"] = df_count.sum(axis=1)

df_count
```

Aus der obigen Tabelle lässt sich ableiten, dass die Methode ``model algo full`` in den meisten Kategorien am besten abgeschlossen hat. Die zweitbeste Methode ist ``model arma-stats``.
