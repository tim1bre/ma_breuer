# analysis of effect of disturbance

- simulations are grouped by disturbance id ``dist_id`` using the mean function

# comparision of simulation with the same disturbance
- in the following section the results of simultions with the same disturbance are compared
- the simulation with the best result leads to incrementing the used strategy
- the following table shows the number disturbances that an approach has performed best on in comparision to the other approaches
- the percentage of won simulations can be also seen
- the ARIMA method has the most "wins" in all categories
- DTW has the second most "wins" in regards to ``res_triang``, performing better than the modelbased approach

```python
# create a list with all used disturances
dist_list = [i for i in range(1, df['dist_id'].max())]

# create a DataFrame with the observed metrices as columns and the strategies as rows
df_best = pd.DataFrame(columns=['res_triang', 'max_loss', 'ttd', 'mttr'])
for idx in df.detect_mode.unique():
    for col in df_best.columns:
        df_best.loc[idx, col] = 0
        
# loop through experiments and determine best strategy
for i in dist_list:
    df_dist = df[df['dist_id'] == i]
    for m in df_best.columns:
        df_dist.sort_values(by=[m], inplace = True)
        df_best.loc[df_dist.iloc[0].detect_mode, m] += 1
        
df_best.sort_values(by=['res_triang'], inplace=True, ascending=False)

# add relative values in %
for col in df_best.columns:
    df_best[col + " in %"] = df_best[col]/df_best[col].sum()*100
    
if export == 1:
    df_best.round(2).to_csv("simulation_winners.csv") # runden und speichern
```
